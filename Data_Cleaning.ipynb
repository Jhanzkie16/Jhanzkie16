```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Jhanzkie16/Jhanzkie16/main/patient_data.csv"
```

```python
df = pd.read_csv(url)
```

```python
pd.set_option("display.max_rows", None)  # Show all rows
print("=== Original Data ===")
print(df)
```

```python
df['Diagnosis'] = df['Diagnosis'].fillna("Not Specified")
```

```python
print("\n=== Data After Filling Missing Diagnosis ===")
print(df)
```

```python
df['Name'] = df['Name'].fillna("Unknown Name")
```

```python
print("\n=== Data After Filling Missing Diagnosis & Name ===")
print(df)
```

```python
avg_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(avg_age)
```

```python
print("\n=== Data After Filling Missing Diagnosis, Name & Age ===")
print(df)
```

```python
df['DischargeDate'] = df['DischargeDate'].fillna("Still Admitted")
```

```python
print("\n=== Data After Cleaning ===")
print(df)
```

```python
duplicates = df[df.duplicated(subset=['Diagnosis'], keep=False)]

print("=== Rows with Duplicate Diagnosis ===")
print(duplicates)
```

```python
df_cleaned = df.drop_duplicates(subset=['Diagnosis'], keep='first')
```

```python
print("=== Data After Dropping Duplicates Based on Diagnosis ===")
print(df_cleaned)
```

```python
df_cleaned.to_csv("patient_data_deduped.csv", index=False)
```
